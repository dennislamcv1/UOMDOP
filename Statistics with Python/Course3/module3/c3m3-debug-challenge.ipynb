```python
import numpy as np 
import scipy.stats as stats
```

```python
# read data from file
heights = np.genfromtxt('data/height_data.csv', delimiter=',')
```

```python
# what's the shape of the data?
heights.shape
```

```python
# print out some summary statistics for the data
print("Mean height: ", np.mean(heights))
print("Median height: ", np.median(heights))
print("Standard deviation: ", np.std(heights))
print("Minimum height: ", np.min(heights))
print("Maximum height: ", np.max(heights))
```

```python
# Calculate how much data is within 2 standard deviation of the mean
mu = np.mean(heights)
sigma = np.std(heights)
lower_bound = mu - 2*sigma
upper_bound = mu + 2*sigma

# What percent of data is within 2 standard deviation of the mean?
prct_within_2_std = np.sum((heights >= lower_bound) & (heights <= upper_bound)) / len(heights)
print(f"Percentage of data within 2 standard deviation of the mean: {prct_within_2_std*100:.2f}%")
```

Wait a second! That doesn't line up very well with our rule of thumb--98% of data should be within _3_ standard deviations of the mean, not 2. What's going on?
