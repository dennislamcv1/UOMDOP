```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

Below is a function, `simulate_one_poll` that simulates conducting a poll on a random sample of a population. Let's assume we're asking them a "yes/no" question, such as "Do you approve of the current president?" 


The function takes the following arguments:

 - `p`: the true proportion of the population that would answer "yes" (1) to the poll question
 - `n_sample`: the number of people to sample from the population
 - `seed`: an optional random seed to ensure reproducibility. By default, an integer seed is provided, meaning the same data will be generated each time. 

The function returns the simulated poll responses as a pandas DataFrame with one column, "response".

```python
def simulate_one_poll(p, n_sample, seed=1):
    responses = stats.bernoulli.rvs(p=p, size=n_sample, random_state=seed)
    responses = pd.DataFrame(responses, columns=["response"])
    return responses
```
