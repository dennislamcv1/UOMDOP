```python
import numpy as np
import matplotlib.pyplot as plt
```

## Review: multiplying vectors
Recall that we can multiply two vectors in a _pairwise_ fashion. Let `a_vector` and `b_vector` be 1 x 5 arrays.

```python
a_vector = np.array([1, 2, 3, 4, 5])
b_vector = np.array([6, 7, 8, 9, 10])
```

Then the pairwise product of the vectors is computed as follows:

```python
a_vector * b_vector
```

## The dot product
Another way to combine two vectors is to take the _dot product_. The dot product is sometimes used to express how much two vectors are "pointing" in the same direction.

How should we think about the direction of vectors? We can picture vectors in a coordinate system. Let's make a few 2-dimensional vectors to practice with (this is for convenience- going into higher dimensions is more challenging to visualize).

```python
c_vector = np.array([1,1])
d_vector = np.array([0.785,1.176])
e_vector = np.array([-1, 1])
f_vector = np.array([-0.785,-1.176])
```

We will define a function for making a blank set of axes for plotting.

```python
def make_axes():
    plt.axvline(0, color='black', lw=0.5)
    plt.axhline(0, color='black', lw=0.5)
    # Scale the plot 
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    fig = plt.gca()
    return fig
```

```python
empty_fig  = make_axes()
```

We'll also define a function for adding a vector to the plot. 

```python
def add_vector_to_figure(figure, v, color='black', label=None):
    origin = np.array([0,0])
    figure.quiver(*origin, *v, scale=1, scale_units='xy', angles='xy', color=color)
    if label is not None:
        figure.annotate(label, v, textcoords="offset points", xytext=(10,10), ha='center')
```

```python
# Add the vector c to the plot we just made in the previous cell
empty_fig  = make_axes()
add_vector_to_figure(empty_fig, c_vector, label = "c")
```

```python
empty_fig  = make_axes()
add_vector_to_figure(empty_fig, c_vector, label = "c")
add_vector_to_figure(empty_fig, d_vector, color='blue', label = "d")
add_vector_to_figure(empty_fig, e_vector, color='red', label = "e")
add_vector_to_figure(empty_fig, f_vector, color='green', label = "f")
```

Looking at the vectors above- which two vectors do we think are most similar? Now let's compute some dot products. 

```python
np.dot(c_vector, d_vector) 
```

```python
np.dot(c_vector, e_vector)
```

```python
np.dot(c_vector, f_vector)
```

Which dot product is bigger? It is largest when the two vectors are pointing in a similar direction! 

```python
np.dot(c_vector, c_vector)
```

Note that the dot product operates on vectors of equal dimensions:

```python
np.dot(a_vector, c_vector)
```
