```python
import scipy.stats as stats
import numpy as np
import seaborn as sns

# Set random seed for reproducibility
rng = np.random.default_rng(2)
```

One kinda of data distribution to know is the _log-normal_ distribution. This distribution has a long right tail. We might see distributions like this when considering data like income, where there is a long tail of a few individuals with very high salaries. 

Although the log-normal distribution is not itself normally distributed, transforming the data by applying a log-transformation converts it to a normal distribution.

Let's generate some data from a log-normal distribution using the `scipy.stats` module's [`lognorm` object](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html) and see how it looks.

```python
mu = 1
sigma = 1
n = 1000

# Generate 1000 random numbers from a lognormal distribution
data = stats.lognorm.rvs(mu, sigma, size = n, random_state = rng)
```

```python
# Plot the distribution of data in Seaborn
sns.histplot(data) 
```

```python
# Calculate some summary statistics of the distribution
mean = np.mean(data)
median = np.median(data)
std = np.std(data)

# Print the summary statistics
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard deviation: {std}")
```
