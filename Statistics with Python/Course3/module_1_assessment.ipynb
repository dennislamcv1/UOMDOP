### 3.1 Using rotation matrices [2 pts]
Recall that we can represent a coordinate $(x,y)$ as a 2-dimensional vector and plot it in a plane. Below is code that will produce a plot of the vector $a$, with coordinates (1,03).

```python
def make_axes():
    plt.axvline(0, color='black', lw=0.5)
    plt.axhline(0, color='black', lw=0.5)
    # Scale the plot 
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    fig = plt.gca()
    return fig

def add_vector_to_figure(figure, v, color='black', label=None):
    origin = np.array([0,0])
    figure.quiver(*origin, *v, scale=1, scale_units='xy', angles='xy', color=color)
    if label is not None:
        figure.annotate(label, v, textcoords="offset points", xytext=(10,10), ha='center')

a = np.array([1, 0])

fig = make_axes()
add_vector_to_figure(fig, a, color='blue', label='a')
```

R = $\begin{bmatrix} cos \theta & -sin \theta \\\ sin \theta & cos \theta \end{bmatrix}$
