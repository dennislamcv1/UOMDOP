```python
import pandas as pd
import numpy as np
```

```python
# create our numpy random number generator
rng = np.random.default_rng(42)
```

```python
# draw values for each person's baseline mean reaction time from a Normal distribution with mean 1 and sd .2
baseline = pd.Series(rng.normal(1, .2, 40))



# draw values for the Before from a Normal distribution with mean from the baseline and sd .2
before = pd.Series(rng.normal(baseline, .2))

# suppose the beer condition increases reaction time by 0.1 seconds
# draw after values from a Normal distribution with mean from the after and sd .2
after = pd.Series(rng.normal(baseline + 0.07, .2))

# assemble a dataframe with the baseline, before and after values, though we'll throw away the baseline for what we show to students.
df = pd.DataFrame({'before': before, 'after': after, 'baseline': baseline})

# save the data to a csv file, excluding the baseline column
df[['before', 'after']].to_csv('data/reaction_times.csv', index=False)
```
