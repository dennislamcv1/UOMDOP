## Separate data by years

Separate data into two groups: 2024 winners, and 2002-2023 winners. The 2024 winners will be used to illustrate `pd.concatenate()`.

```python
filepath = "./mega_millions-source-20240518.csv"
data = pd.read_csv(filepath, dtype=str)
```

```python
data = data.astype(str)  # convert all columns to string
data.info()
```

```python
# Add leading zero
# WARN: Despite passing dtype=str to read_csv() the lambda encountered a float
# Resolved by casting all column values to strings explicitly: data = data.astype(str)
data["Multiplier"] = data.loc[:, "Multiplier"].apply(lambda x: x.zfill(2) if x != "nan" else "")
data["Mega Ball"] = data.loc[:, "Mega Ball"].apply(lambda x: x.zfill(2))
```

## Filtering function

```python
def match_year(draw_date, start, end):
    """TODO"""

    for i in range(start, end + 1):
        if int(draw_date[-4:]) == i:
            return True
    return False
```

## Get 2023-2024 winners

```python
# Get 2023-2024 data (don't convert date, work with date string)
mask = data["Draw Date"].apply(match_year, start=2023, end=2024)
# mask = data["Draw Date"].apply(lambda x: x.endswith("2023") | x.endswith("2024")) # Alternative
# mask = data["Draw Date"].str.endswith("2023") | data["Draw Date"].str.endswith("2024") # Alternative
data_2023_24 = data[mask]
data_2023_24.info()
```

```python
# Write to file
filepath = "./mega_millions-2023_24.csv"
data_2023_24.to_csv(filepath, index=False)
```

## Get 2016-2022 winners

```python
# Get 2016-2022 data (don't convert date, work with date string)
mask = data["Draw Date"].apply(match_year, start=2016, end=2022)
data_2016_22 = data[mask]
data_2016_22.info()
```
