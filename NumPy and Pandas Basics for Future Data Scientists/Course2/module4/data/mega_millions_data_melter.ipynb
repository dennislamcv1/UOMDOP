```python
import pandas as pd
```

## Retrieve data

```python
data = pd.read_csv("./mega_millions_pick5_cols-2002_24.csv")
```

```python
data.head(3)
```

## Convert Draw Date

```python
data.loc[:, "draw_date"] = pd.to_datetime(data.loc[:, "draw_date"])
```

## Melt Data

Unpivot from wide to long format

```python
columns = data.columns.tolist()
data_m = data.melt(
    id_vars=["draw_date"], value_vars=columns[1:], var_name="variable", value_name="value"
).sort_values(by=["draw_date", "variable"], ascending=[False, True], ignore_index=True)

data_m.head(24)
```

```python
data_m.shape
```

## Convert datetime back to original string format

Gotcha: [`astype()` and `loc()` behaviors](https://pandas.pydata.org/docs/user_guide/basics.html#astype)

> When trying to convert a subset of columns to a specified type using astype() and loc(), upcasting occurs.
>
> loc() tries to fit in what we are assigning to the current dtypes, while [] will overwrite them taking the dtype from the right hand side. Therefore the following piece of code produces the unintended result.

```python
data_m["draw_date"] = data_m["draw_date"].dt.strftime("%m/%d/%Y")  # works
# data_m.loc[:, "draw_date"] = data_m.loc["draw_date"].dt.strftime("%m/%d/%Y")  # fails due to loc() upcasting
data_m.dtypes
```
