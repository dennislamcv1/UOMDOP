## Reading CSV DataFrames

Pandas provides several functions to read and write DataFrames from and to different file formats. One of the most commonly used formats is CSV (Comma Separated Values). Let's start by reading a DataFrame from a CSV file.

We'll be using a dataset showing the winning numbers in the New York state powerball lottery. It's downloaded from a [government data catalog](https://www.google.com/url?q=https://catalog.data.gov/dataset/lottery-powerball-winning-numbers-beginning-2010)

I encourage you to start by looking at the dataset as a .csv file. It's in sample_data/powerball_winners.csv

```python
# Read a CSV file
powerball_df = pd.read_csv('sample_data/powerball_winners.csv')
```

## Converting a column to datetime objects

If we want to treat the data as a time series, we will want to use python's built-in handlers for date times. That lets us do things like compute the elapsed time between two dates. So, let's convert the Draw Date column to be datetime objects.

```python
powerball_df.dtypes
```

```python
pd.to_datetime(powerball_df['Draw Date'])
```

```python
powerball_df['Draw Date'] = pd.to_datetime(powerball_df['Draw Date'])
powerball_df.dtypes
```

```python
powerball_df.head(3)
```

```python
# how long is the span of this dataset?
elapsed_time = powerball_df['Draw Date'].iloc[-1] - powerball_df['Draw Date'].iloc[0]
elapsed_time
```

## Indexes
When reading a CSV file, pandas automatically assigns integer index labels to the rows.

But we can change what's used as the index.

```python
powerball_df.index
```

```python
powerball_df.loc[0]
```

```python
powerball_df.iloc[0]
```

```python
# make a copy of the dataframe with the 'Draw Date' as the index
reindexed_df = powerball_df.set_index('Draw Date')
```
