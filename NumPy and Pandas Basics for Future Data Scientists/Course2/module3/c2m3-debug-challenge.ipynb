```python
import pandas as pd
import numpy as np
```

This is a dataset of the patients who are seen by a particular doctor's office, and the number of COVID-19 vaccinations they have received. We want to understand how many patients are awaiting vaccinations so we can make sure we have enough doses on hand. 


```python
df = pd.read_csv('data/patient_vaccination_status.csv')
df.head()
```

Let's create a categorial variable for different age groups, binned into decades. This will be helpful later on if we want to understand if there are any age groups being underserved.

```python
bins = np.arange(0, 101, 10)
labels = [f'{i}-{i+9}' for i in range(0, 100, 10)]
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
df.head()
```

How many patients have never been vaccinated?

```python
# count number of rows where n_doses == 0
n_not_vaccinated = df[df['n_doses'] == 0].shape[0]
print(f'Number of patients never vaccinated: {n_not_vaccinated}')
```

The COVID-19 vaccination typically requires multiple doses, depending on the specific vaccine given. For this dataset, let's assume that 2 doses is considered fully vaccinated. How many patients are fully vaccinated? How many are partially vaccinated?


```python
any_vaccine = df[df['n_doses'] != 0]
n_any_vaccine = any_vaccine.shape[0]
# Print out how many patients have received any vaccine
print(f'Number of patients who have received at least one dose: {n_any_vaccine}')

any_vaccine.n_doses.value_counts()
```
