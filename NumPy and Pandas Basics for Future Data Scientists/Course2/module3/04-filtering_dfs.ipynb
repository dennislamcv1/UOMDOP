## Filtering Data
We have previously seen how to select subsets of rows and columns by specifying column names or index values, or slices of them.

It's also very common to want to filter the data based on criteria: only keep rows satisfying certain conditions.

Let's start with a simple DataFrame.

```python
# Create a simple DataFrame
df = pd.DataFrame({
    'Name': ['John', 'Soo', 'Amit', 'Linda', 'Sam'],
    'Age': [28, 24, 42, 60, 26],
    'City': ['Ann Arbor', 'Seoul', 'Chennai', 'Ann Arbor', 'Paris'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'NonBinary']
})
df
```

Filtering based on numeric criteria

```python
df['Age'] > 40
```

```python
# select the people who are older than 40
df.loc[df['Age'] > 40]
```

Filtering based on categorical variables

```python
# You will also see other people write code that looks like this, but we will try to use .loc
df[df['Age'] > 40]
```

```python
# select the people who live in Ann Arbor
df.loc[df['City'] == 'Ann Arbor']
```

## Categorical variables


Categorical variables are variables that can take on one of a limited number of values, the categories.
Traditionally, in surveys 'Gender' has been treated as a binary category, with 'Male' and 'Female' as the only valid values. More recently, some surveys have included additonal gender options, or sometimes they just include one additional extra value, 'NonBinary'.

```python
# Filter the DataFrame to select only the rows for females
df.loc[df['Gender'] == 'Female']
```

Right now, the DataFrame does not enforce the constraint that 'Gender' is a categorical variable. I could add another row where the Gender column is "Fish".


```python
# Add a row to the DataFrame with Gender set to "fish"
df.loc[len(df)] = ['Dory', 2, 'Great Barrier Reef', 'Fish']
df
```

We can set a constraint on Gender so that it only accepts certain values.

```python
from pandas.api.types import CategoricalDtype

# Define the categories
categories = ['Male', 'Female', 'NonBinary']

# Create the categorical data type
genders = CategoricalDtype(categories=categories, ordered=True)

# Convert the column to the categorical data type
df['Gender'] = df['Gender'].astype(genders)
df
```

Notice that Dory has the value NaN (the pandas equivalent of None in base python). When we restricted the Gender column to only allow three values, and "Fish" wasn't one of them, it converted "Fish" to Nan.

## Combining filter conditions

We can combine filter conditions using logical operators, and, or, and not.


```python
# only people over 40 who live in Ann Arbor
df.loc[(df['Age'] > 40) & (df['City'] == 'Ann Arbor')]
```

```python
# only people who are either older than 40 *or* live in Ann Arbor
df.loc[(df['Age'] > 40) | (df['City'] == 'Ann Arbor')]
```

```python
# only people who are younger than 40 and do *not* live in Ann Arbor
df.loc[(df['Age'] < 40) & ~(df['City'] == 'Ann Arbor')]
```
