# Introduction to Pandas

Pandas is a powerful data manipulation library in Python. It's built on top of the NumPy package, meaning a lot of the structure of NumPy is used or replicated in Pandas. However, Pandas provides more functionality and flexibility for handling data than NumPy. In this notebook, we will introduce two fundamental pandas objects: DataFrames and Series, and compare them with NumPy arrays.

## Series

A Series is a one-dimensional labeled array. It's similar to a NumPy 1D array, but with added features:

- **Index**: Each element in a Series has an associated label, called an index. This makes it easier to refer to specific data points, similar to how you can refer to columns by name in a DataFrame.
- **Additional data types**: While numpy arrays can only contain integers, floats, or boolean values, a Series can hold any type of python object. Some of the commonly used types include:
    - `datetime`, for working with time series data
    - `category`, a pandas-specific type, for working with data that can only consist of one of a set of discrete values


[Pandas intro documentation for Series](https://pandas.pydata.org/pandas-docs/version/2.2/user_guide/dsintro.html#series).

```python
# Creating a Series
s = pd.Series([1, 3, 5, 11.0, 6, 8])
s
```

```python
# Setting index labels for the series
s.index = ['Person 1', 'Person 2', 'Person 3', 'Person 4', 'Person 5', 'Person 6']
s
```

## Accessing Elements in a Series
Now that we have string labels for our index, we can distinguish between accessing elements using `.loc` (which uses the index labels) and `.iloc` (which uses integer locations).

```python
# Accessing a single element using .loc (index labels)
s.loc['Person 2']
```

```python
# Accessing a single element using .iloc (positonal index)
s.iloc[0]  # Accessing the first element of the series
```

```python
# Taking a slice, using .loc
# Note that the slice is inclusive of the last element, unlike numeric indexing in Python
s.loc['Person 2':'Person 5']
```

```python
# Taking a slice, using .iloc
# Note that the slice is **exclusive** of the last element, like all numeric indexing in Python
s.iloc[0:3]
```

You can also index or slice using square brackets, as with python lists. This has the same effect as using .iloc. But it's better to get used to using .iloc and, even more commonly, .loc.

```python
s[0]
```

```python
s[:2]
```

## Setting values using .loc and .iloc
We can also set the value of a specific cell using `.loc` and `.iloc`.

```python
s.loc['Person 1'] = -50
s
```

```python
s.iloc[0] = 7000  # Setting the first element of the first row
s
```

```python
# You can do this, but get in the habit of using .iloc (or better yet, .loc), for reasons that will be discussed later.
s[0] = 5000
s
```

## DataFrame

A DataFrame is a two-dimensional tabular data structure with labeled axes (rows and columns). It's similar to a NumPy 2D array, but with added features:

- **Named columns**: Each column in a DataFrame has a name, which makes it easier to refer to specific columns in a way that is easy to read and less error-prone (you won't get the wrong column by accident!)
- **Labeled rows**: Each row has an "index" value, which can be a number or a string.
- **Heterogeneous data types**: Unlike NumPy arrays, which require all elements to be of the same data type, DataFrames can have different data types for different columns.
- **Additional data types**, as described for Series.

[Pandas intro documentation for DataFrames](https://pandas.pydata.org/pandas-docs/version/2.2/user_guide/dsintro.html#dataframe).


```python
# Creating a DataFrame
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df = pd.DataFrame(data)
df
```

Pandas provides several ways to extract and set data in a DataFrame. This flexibility makes it easy to perform complex data manipulations.

Let's start by extracting data. You can extract data from a DataFrame using column names, row labels, or their integer locations.

```python
# Extracting a single column by name
df['Name']
```

Each column is a pandas Series object.

```python
type(df['Name'])
```

We can also extract a single column using .dot notation

```python
df.Age
```

```python
# Extracting a DataFrame with multiple selected columns
df[['Name', 'City']]
```

Let's add an index for the rows of the DataFrame to make it easier to access rows by label

```python
df.index = ['Person 1', 'Person 2', 'Person 3', 'Person 4']
df
```

```python
# Extracting rows by label through loc
df.loc['Person 1']
```

```python
# Extracting rows by integer location through iloc
df.iloc[0]  # Extracts the first row
```

```python
df
```

```python
# Accessing a cell
df.loc['Person 1', 'Age']
```

You can also access a cell through "chained indexing", but this is discouraged. Some things will be changing between version 2 and version 3 of pandas. To not get hung up by those, you're better off just getting in the habit of using .loc.

More details can be found in [the docs](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#why-does-assignment-fail-when-using-chained-indexing), if you want to try to understand the gory details.

```python
# Works, but discouraged
df['Age']['Person 1']
```

## Modifying the data
Now let's look at setting data. You can modify the data in a DataFrame, add new columns, or change the values of existing columns, using all the same syntax we had for accessing them.

```python
# Modifying a cell
df.loc['Person 1', 'Age'] += 100
df
```

```python
df['Name'].str.upper()
```

```python
# Modifying a whole column
df['Name'] = df['Name'].str.upper()
df
```

```python
# Adding a new column
df['Country'] = ['USA', 'France', 'Germany', 'UK']
df
```

```python
# This syntax doesn't work for adding a new column
df.foo = [1, 2, 3, 4]
df
```

## Adding and Dropping Columns

One of the advantages of pandas over NumPy is the ability to add and drop columns from a DataFrame, by name. This can be useful if you have a dataset with many columns and you want to focus on a subset of the columns. You could do this with slicing and joining in NumPy, but it's a lot easier to write and read the code in Pandas.

```python
df
```

```python
# Adding a new column
df['Profession'] = ['Engineer', 'Doctor', 'Artist', 'Scientist']
df
```

```python
# Dropping a column
df = df.drop('Age', axis=1)
df
```

```python
# Why doesn't this drop the 'Profession' column?
df.drop('Profession', axis=1)
df
```

```python
df
```

```python
# Dropping a row
df = df.drop('Person 1', axis=0) # axis=0 is the default but specifying it here for clarity
df
```

```python
df
```

```python
df.iloc[1].name
```

```python
df = df.drop(df.iloc[1].name)  # Dropping the second row
df
```
