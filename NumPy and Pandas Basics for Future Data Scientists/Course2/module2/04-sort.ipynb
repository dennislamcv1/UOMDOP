## Sorting vectors

Let's make a 1-dimensional array of randomly generated numbers. (Each number is between 0 and 9).

```python
arr = np.random.randint(10,size=14)
arr
```

What if we wanted to sort this array from smallest to largest? We can use the [array method `.sort()` ](https://numpy.org/doc/1.26/reference/generated/numpy.ndarray.sort.html#numpy.ndarray.sort)  for this. This is analogous to the `.sort()` method for python lists.

```python
arr.sort()
arr
```

Notice what happened - the variable `arr` has been sorted _in place_! What if we had wanted to return a sorted _copy_ of the original array? We can use the function `np.sort()` for this. This is analogous to the `sorted` function for python lists.

```python
# Let's make a new randomly generated array to practice on.
arr_b = np.random.randint(10, size =12)
arr_b
```

```python
sorted_arr_b = np.sort(arr_b)
sorted_arr_b
```

And let's confirm that `arr_b` hasn't changed.

```python
arr_b
```

We can also sort in descending order- there are a few ways to do this, one is with the `np.flip()` function.

```python
# This one doesn't work; np.sort isn't the same as .sort in base python...
np.sort(arr_b, reverse=True)
```

```python
# this works
np.flip(np.sort(arr_b))
```

Another trick in Python is to use something called the ["extended slicing"](https://docs.python.org/3/whatsnew/2.3.html#extended-slices). This is a way of indexing an array that returns the elements in reversed order.

Here we take all the elements, but we increment the index by -1 each time, meaning we go from back to front.

```python
# or this
sorted_arr_b[::-1]
```

## Sorting matrices
When we have more than 1 dimension that we could possibly sort along, we have to pay extra attention to how the sorting is implemented.

Let's practice with a 4-by-4 matrix of randomly generated numbers.

```python
matrix_a = np.random.randint(10,size = (4,4))
matrix_a
```

What happens if we simply try the `.sort()` method with its default settings?

```python
matrix_a_sort_default = np.sort(matrix_a)
matrix_a_sort_default
```

This sorted every _row_ of our matrix! By default, `np.sort()` sorts "along" the *last* axis of the array, in this case the columns. Sorting "along" the columns means to hold the row fixed and vary the column index.

We can also use the `axis` argument to `np.sort()` to specify which axis we want to sort by.

```python
# same as default
matrix_a_sort_axis = np.sort(matrix_a, axis =  -1)
matrix_a_sort_axis
```

```python
# sort "along" the rows, meaning hold the column fixed and sort the values from the different rows (i.e., sort within each column)
matrix_a_sort_axis = np.sort(matrix_a, axis = 0)
matrix_a_sort_axis
```
