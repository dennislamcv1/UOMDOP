## Changing the shape of an array, while keeping the contents

We can chop up a 1-dimensional array into a bunch of rows, or even make it into a higher dimensional tensor.

First we will create a one-dimensional array.

```python
arr = np.array([1, 2, 3, 4, 5, 6])
```

Let's reshape it so that it has two columns and three rows.

```python
reshaped_arr = arr.reshape(3, 2)
reshaped_arr
```

If you specify -1 for the size of one of the dimensions, numpy will figure it out.

```python
# For example, we can just specify three rows.
arr.reshape(3, -1)
```

```python
# Or just specify two columns
arr.reshape(-1, 2)
```

We could also reshape the original array so it has three columns.

```python
arr.reshape((2, 3))
```

## Inverse: Flattening a multi-dimensional array


```python
reshaped_arr = arr.reshape(3, 2)
flattened = reshaped_arr.flatten()
flattened
```

Because these numpy methods return array objects, we can chain together operations.

```python
arr.reshape(3,2).flatten()
```

You can also flatten using .reshape

```python
reshaped_arr.reshape(6)
```

# Does reshape mutate the array or make a copy?
What happens if we run the reshape method without storing the result in a new variable?

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped_arr = arr.reshape(3, 2)
reshaped_arr
```

Question: did we change the shape of `arr`?

```python
arr.shape
```

Does modifying the contents of reshaped_arr mutate the original arr?

```python
reshaped_arr[0,0] = -50
reshaped_arr
```

```python
arr
```

## Lesson
Numpy will *try* to reshape the array with a "view" rather than making a complete copy. If you modify the contents of the reshaped array, you will have modified the contents of the original array as well. That can be confusing.

Look out for this possibility when you are debugging. How did the contents of this array change during my code execution, even though the code never refers to that array by its original name? Maybe it's because some view of that array, created from reshaping or slicing, got modified.

To make things worse, if it can't reshape as a view, it *will* make a complete copy. So you can't always count on modifications to the reshaped array propagating back to the original array.

What about when you flatten? Do you get a copy?

```python
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
arr2
```

```python
flattened = arr2.flatten()
flattened
```

```python
flattened[0] = -100
flattened
```

```python
arr2
```

```python
flattened_with_reshape = arr2.reshape(-1)
flattened_with_reshape
```

```python
flattened_with_reshape[0] = -100
flattened_with_reshape
```

```python
arr2
```

## Lesson
`.flatten()` always makes a copy of an array, never a view. If you modify the flattened copy, it doesn't affect the original.

So, we have two ways to flatten an array. If you use reshape, it might create a view rather than a copy, which is more efficient, but might lead to confusion if you modify the flattened copy.
