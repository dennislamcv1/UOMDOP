Broadcasting is a  feature in NumPy that lets you do arithmetic operations on arrays with different shapes.


## Broadcasting with a scalar

A _scalar_ is a 0-dimensional array- which is less scary than it sounds. For example, the integer 10 is a scalar!

If we want to add the number 10 to every element of a `numpy` array, broadcasting lets us do this without writing a for-loop, or creating a full matrix of 10s.

```python
arr = np.array([1, 2, 3, 4, 5])
arr
```

Now let's add the scalar 10 to this array.

## With broadcasting
Instead of explicitly creating the second array, we can provide a scalar and numpy will implicitly treat it as the second value to combine with each element of the first array.

This allows for a more efficient execution, because the python interpreter never has to actually create the full second array; it just broadcasts values as needed. It's a little analogous to the difference between lists in base python and generators, which produce a sequence of values on demand.

```python
result = arr + 10
result
```

Even though our array `arr` and our scalar 10 have different shapes....

```python
arr.shape
```

```python
np.shape(10)
```

...we are able to _broadcast_ 10 to every element of `arr`.

## What is happenening when we broadcast?
Let's review a helpful picture from the [broadcasting docs](https://numpy.org/doc/1.26/user/basics.broadcasting.html).

Here's a conceptual diagram of what is happening when we do an operation with two differently sized arrays (here, a 1-dimensional array $a$ and a scalar $b$):

![Broadcasting_1](https://numpy.org/doc/1.26/_images/broadcasting_1.png)



We can imagine _stretching_ the scalar $b$ into an array, so it becomes the same shape as $a$.  This corresponds to the code below:

```python
a = np.array([1,2,3])
b = 2
a*b
```

**True or false:** When we do an operation on two differently sized arrays with broadcasting, the result will always be the same size as the larger array.

We can also visualize broadcasting with higher-dimension arrays.

![broadcasting_2](https://numpy.org/doc/1.26/_images/broadcasting_2.png)

This corresponds to the following code:

```python
a = np.array([[ 0,  0,  0],
              [10, 10, 10],
              [20, 20, 20],
              [30, 30, 30]])
b = np.array([1, 2, 3])
a + b
```

We can even broadcast by stretching both arrays, on different dimensions.

![broadcasting_2](https://numpy.org/doc/1.26/_images/broadcasting_4.png)

Make the 4x1 array

```python
a = np.array([[0], 
              [10], 
              [20], 
              [30]])
a
```

```python
# alternative way to make the 4x1 array
a = np.array([0, 10, 20, 30]) # make a 1-d array with four 4 elements
a = a[:, np.newaxis] # add an extra axis to it
a
```

Make the 1x3 array

```python
b = np.array([1, 2, 3])
b
```

```python
c = a + b
c
```
