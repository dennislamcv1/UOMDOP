## Joining arrays

The numpy documentation uses the term join generically to refer to a variety of operations that put together the contents of two arrays to make a bigger array. Don't get confused with the database join operaiton on two tables, like in SQL; it's not the same as that.



```python
# Let's make some arrays to practice joining.
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

```python
arr1
```

```python
arr2
```

Let's try the function `np.concatenate()` and see what happens.

```python
np.concatenate((arr1, arr2))
```

We can also use the functions `np.hstack()` and `np.vstack()`, to join (stack) the arrays horizontally or vertically.

```python
np.hstack((arr1, arr2))
```

```python
np.vstack((arr1, arr2))
```

Let's see some examples with higher-dimension arrays:

```python
arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[5, 6], [7, 8]])
```

```python
arr3
```

```python
arr4
```

```python
np.concatenate((arr3, arr4))
```

When multiple axes are present, we can also use the `axis` argument to specific which one to join along:

```python
np.concatenate((arr3, arr4), axis=0)
```

And we have `np.vstack()` and `np.hstack()`:

```python
np.vstack((arr3, arr4))
```

```python
np.hstack((arr3, arr4))
```

Arrays do not have to be exactly the same size to be joined!

```python
arr4 = np.array([[5, 6], [7, 8]])
```

```python
arr5 = np.array([9,10])
```

```python
np.vstack((arr4, arr5))
```

But they do need to be compatible.

```python
arr6 = np.array([9,10,11])
```

```python
np.vstack((arr4, arr6))
```
