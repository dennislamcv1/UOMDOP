```python
import numpy as np
```

Let's make a few different arrays, and then we'll inspect them.

```python
arr_a = np.array([[1, 2, 3],[4, 5, 6]])
arr_b = np.array([1.5, 2.7, 3.9])
arr_c = np.array([True,False,True])
```

By design, every `numpy` array has certain _attributes_. You can think of each as a property that particular array possesses. Let's look at three of them:

*   Size
*   Shape
*   Data-type



```python
arr_a.size
```

```python
arr_a.shape
```

```python
arr_a.dtype
```
