```python
import numpy as np
```

Sometimes, we are missing data! This happens _a lot_ when we're collecting data. It's important to represent this clearly in our data. One way to do this is with a "NaN"- short for "not a number". `numpy` has a special way of representing this. Let's work with an example.

```python
arr = np.array([[1, 2, np.nan],
                [4, np.nan, 6],
                [7, 8, 9]])
```

A helpful `numpy` function for handling arrays with NaN values is `isnan()`.

```python
np.isnan(arr)
```

## Counting NaN values

See how the output compares to what's in our array. What do you notice? Next, let's see how we can count the number of NaNs. We can use another `numpy` function called `count_nonzero()`.

```python
np.count_nonzero(np.isnan(arr))
```

Another way we can get the same answer is through a sum, because Boolean data in Python can be added. Just like this:

```python
True + True + True + False
```

So if we have an array of Boolean values (representing whether a NaN value is present at that position in our array) we can add them all up.

```python
np.isnan(arr).sum()
```

## Working with NaN values

Some functions will return NaN if there are _any_ NaN values in the array being operated on. For example, what if we tried to use the `.sum()` function on our array that contains NaNs?

```python
arr.sum()
```

There are often versions of common functions in `numpy` that handle NaNs, like this:

```python
np.nansum(arr)
```

Notice something interesting- `arr.sum()` works _on_ the object `arr`. But `np.nansum()` is a function that takes an argument, `arr`.

Now, depending on how the NaNs got there, we might want to replace them with a particular value. That can save us some headaches when it's time to do arithmetic! There are lots of ways to choose the value you'll use, but in this case, let's use 0.

```python
arr[np.isnan(arr)] = 0
```

Another way to do this is with a `numpy` function called `where()`. The syntax takes a moment to understand because it's a compact represntation of a control-flow statement.

```python
np.where(np.isnan(arr), 0, arr)
```
