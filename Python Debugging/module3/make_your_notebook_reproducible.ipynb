# Make your notebook reproducible
One of the benefits of using a Jupyter notebook is that you can hand off your work to someone else (or Future You, when you come back to your work after a few days off!) and quickly replicate your computations and visualizations. But after you've been working on a problem for a while, your notebook might start looking like a scratchpad than a finished report. That's natural and an inevitable part of the process for most people. Let's see a few ways to tidy up a notebook so it's ready to be reproduced. Here's an example notebook that is still in "play" mode.

```python
import random

# Choose a random element from a list
fruit_list = ['apple', 'banana', 'orange', 'pineapple', 'melon', 'orange', 'pineapple']
prize = random.choice(fruit_list)
print("You've won:", prize)
```

```python
winning_odds = my_function(fruit_list)

# find the fruits with the highest odds of winning
x = max(winning_odds.values())
y = [key for key, value in winning_odds.items() if value == x]
y
```

```python
# create a function 
def my_function(lst):
    odds = {}
    total_elements = len(lst)
    # this is a for loop
    for element in lst:
        count = lst.count(element)
        odds[element] = count / total_elements
    return odds

# don't use this!!!!!
#def my_function(lst):
#    odds = []
#    total_elements = len(lst)
#    for element in lst:
#        count = lst.count(element)
#        odds.append(count / total_elements)
#    return odds
```
