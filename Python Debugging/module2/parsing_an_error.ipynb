## Let's see some tracebacks!
When we read an error message (also called a "traceback" ) we want to extract two important pieces of information from it: 

- What TYPE of error is this? 
- WHERE is the error? We're looking for a line number!

Reading tracebacks from _bottom to top_ will usually help you find this info efficiently. Let's see some examples together.
