### Code Block 2

```python
def convert_to_secs(track_length):
    """Converts track length from minutes:seconds to seconds.

    Parameters:
        track_length (str): The track length in minutes:seconds format.

    Returns:
        int: The track length in seconds.
    """

    mins_secs = track_length.split(":")
    secs = int(mins_secs[0]) * 60 + int(mins_secs[1])
    return secs

track_length = "4:30"
secs = convert_to_secs(track_length)
print(f"Track length in seconds: {secs}")
```
