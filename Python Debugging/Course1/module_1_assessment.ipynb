Below is a function `longer_than_n()` that takes as arguments a list of strings and an integer threshold, `n`. It returns a list of only the strings whose length exceeds the threshold. Rewrite this function using a list comprehension and call it `refactored_longer_than_n()`.

```python
def longer_than_n(strings, n):
    long_strings = []
    for string in strings:
        if len(string) > n:
            long_strings.append(string)
    return long_strings


# YOUR CODE HERE
raise NotImplementedError()
```
