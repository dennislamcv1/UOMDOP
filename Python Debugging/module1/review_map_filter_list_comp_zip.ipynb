## List Comprehensions, map and filter

Recall the basic structure of a list comprehension expression:
- `[<transformer_expression> for <loop_var> in \<sequence\> if <filtration_expression>]`
- The last element, the filtration expression, is optional

To understand a list comprehension, I recommend reading the parts in the following order:
- The sequence: what are we iterating through?
- The loop variable: what is the variable name that will be used?
- The filtration expression: if an item doesn't pass the filter, the transformer expression won't be executed for it, so understand this first
- The transformer expression

An example with transformer expression only

```python
def doubler(lst: list[int | float]) -> list[int | float]:
    return [2 * num for num in lst]
```

```python
doubler([2, 5, 7.3])
```

```python
doubler(range(10))
```

### `assert` statements and f-strings
`assert` is followed by a boolean expression
- if the expression evaluates to `True`, nothing happens
- if the expression evaluates to `False`, an Exception is raised

The purpose of `assert` is to record the programmer's expectations about the state of a computation in a way that can be automatically checked. If the programmer's expectations are violated, an error will occur immediately. This can be useful at helping to identify problems that would otherwise be hard to debug.

f-strings are a way to produce a text string with variable values interpolated in the middle of them. It's a more modern version of what was once accomplished using the `.format()` method.

```python
x = 10
for _ in range(20):
    x -= 1
```

```python
assert (
    x >= 0
), f"I thought x should be positive at this point but it's actually {x}; something must be wrong."
```

```python
assert doubler([4, 11, 6, 8, 23]) == [8, 22, 12, 16, 46]
```

```python
assert doubler([4, 11, 6, 8, 23]) == [8, 22, 12, 16, 46, 100]
```

### Alternative with map
`map` is a function that takes two inputs:
- a transformer function
- a sequence

The transformer function takes one item as input and returns a transformed version of it. The transformer is called on each of the items in the sequence, and the results are accumulated into the result. The result is returned as an "iterator", which we can convert into a list if we want to.

We can use it to do the same thing we can with a list comprehension. It's more common to use list comprehensions, but it's helpful for you to understand how to do the same thing using `map`, because that is conceptually similar to the `apply` method that we'll see later in pandas, where some transformer function is applied to each row each column in a table. Understanding `map` will help get you ready for that.


```python
def doubler(lst: list[int | float]) -> list[int | float]:
    return [2 * num for num in lst]
```

```python
def doubler(lst):
    def num_doubled(num):
        return 2 * num

    return list(
        map(num_doubled, lst)
    )  # map returns an iterator; we explicitly convert that to a list
```

```python
assert doubler([4, 11, 6, 8, 23]) == [8, 22, 12, 16, 46]
```

### Alternative with an anonymous (unnamed) `lambda` expression
A lambda expression creates a function without giving it a name.

Unlike `def`, which is a statement that always has to begin a line of code, a `lambda` expression can occur anywhere in code that a function is expected.

The syntax of a lambda expression is:
- `lambda <param names>: <expression referring to param names>`
    - Note that we don't explicitly have a `return` statement. The value of the expression after the colon will automatically be returned

```python
def doubler(lst):
    def num_doubled(num):
        return 2 * num

    return list(
        map(num_doubled, lst)
    )  # map returns an iterator; we explicitly convert that to a list
```

```python
def doubler(lst):
    return list(
        map(lambda num: 2 * num, lst)
    )  # map returns an iterator; we explicitly convert that to a list
```

```python
assert doubler([4, 11, 6, 8, 23]) == [8, 22, 12, 16, 46]
```

## A list comprehension example with filtration expression only

```python
def snames(strings: list[str]) -> list[str]:
    return [s for s in strings if s[0] == "S"]
```

```python
snames(["Strong", "smart", "fast"])
```

```python
firstnames = ["Jack", "Samantha", "Nelson", "Swathi", "Shu-fen"]
lastnames = ["Vikander", "Anderson", "Tyler", "Anand", "Li"]
```

```python
assert snames(firstnames) == ["Samantha", "Swathi", "Shu-fen"]
assert snames(lastnames) == []
```

### Alternative using `filter`
Instead of using a filtration expression in a list comprehension, we can use the `filter` function.  Learning to use this alternative will set you up nicely for filtering numpy arrays and pandas dataFrames when you learn about them.

`filter` takes two inputs:
- a filtration function
- a sequence

The filtration function should take one item as an input and return a Boolean, `True` or `False`.

When `filter` is executed, the filtration function is applied to each item. Those items for which the filtration function returns `True` are included in the final list; the others are not (they are "filtered out")

```python
def snames(strings: list[str]) -> list[str]:
    return [s for s in strings if s[0] == "S"]
```

```python
def snames(strings):
    def sname(s):
        return s[0] == "S"

    return list(filter(sname, strings))
```

```python
assert snames(firstnames) == ["Samantha", "Swathi", "Shu-fen"]
assert snames(lastnames) == []
```

# A List Comprehension with Both Elements

Below, we have provided a list of tuples that contain students’ names and their final grades in PYTHON 101. 

Using a list comprehension, let's create a new list that contains the *just the names* of students who passed the class (had a final grade of 70 or greater) and assign it to the variable `passing_students`.

```python
# names generated at https://www.name-generator.org.uk
scores = [
    ("Sirisha Patel", 95),
    ("Magan Sharma", 63),
    ("Suchitra Mohammad", 100),
    ("Nico Kaur", 50),
    ("Tarala Aziz", 75),
]
```

```python
passing_students = [name for (name, score) in scores if score >= 70]
```

```python
passing_students
```

```python
assert passing_students == ["Sirisha Patel", "Suchitra Mohammad", "Tarala Aziz"]
```

Alternative with `map` and `filter`. We first apply the filter, and then the map operation, because the transformer expression should only be run on items that pass the filter. In this case, it doesn't really matter which order we do it in, but we could have a list where the transformer expression yields an error on items that are filtered out, so it's best practice to filter first.

Unfortunately, we can't automatically unpack each tuple into two parameter names in a lambda expression, so the code is harder to read

```python
passing_students = list(map(lambda tup: tup[0], filter(lambda tup: tup[1] >= 70, scores)))
```

```python
assert passing_students == ["Sirisha Patel", "Suchitra Mohammad", "Tarala Aziz"]
```

## Zip

`zip` takes two or more sequences of the same length and accumulates one tuple for each position. The tuple for the first position has the first element of each of the sequences. The second tuple has the second element of each of the sequences. And so on.

`zip` returns an iterator, which can be turned into a list if needed.


```python
countries = [
    "Canada",
    "Mexico",
    "United States",
    "Guatemala",
    "Nicaragua",
    "Honduras",
    "El Salvador",
    "Costa Rica",
    "Panama",
]
codes = [1, 52, 1, 502, 505, 504, 503, 506, 507]
```

```python
country_codes = list(zip(countries, codes))
country_codes
```

```python
assert country_codes[3] == ("Guatemala", 502)
```

Here we are filtering the output of `zip` to produce a list that contains the names of species whose populations are below 2500.

```python
species = [
    "golden retriever",
    "white tailed deer",
    "black rhino",
    "brown squirrel",
    "field mouse",
    "orangutan",
    "sumatran elephant",
    "rainbow trout",
    "black bear",
    "blue whale",
    "water moccasin",
    "giant panda",
    "green turtle",
    "blue jay",
    "japanese beetle",
]

populations = [
    10000,
    90000,
    1000,
    2000000,
    500000,
    500,
    1200,
    8000,
    12000,
    2300,
    7500,
    100,
    1800,
    9500,
    125000,
]
```

```python
endangered = [animal for animal, pop in zip(species, populations) if pop < 2500]
```

```python
assert endangered == [
    "black rhino",
    "orangutan",
    "sumatran elephant",
    "blue whale",
    "giant panda",
    "green turtle",
]
```

```python
endangered = list(
    map(lambda tup: tup[0], filter(lambda tup: tup[1] < 2500, zip(species, populations)))
)
```

```python
assert endangered == [
    "black rhino",
    "orangutan",
    "sumatran elephant",
    "blue whale",
    "giant panda",
    "green turtle",
]
```
