```python
def doubler(lst):
    def num_doubled(num):
        return 2 * num

    return list(map(num_doubled, lst))
```

**Example explanation**: This code defines a function `doubler()` that accepts a sequence `lst` and maps the function `num_doubled()` over each element of `lst`. It returns a list that has all the numbers in the original list doubled.

```python
def is_dot_org(strings):
    def check_domain_extension(s):
        return s.split(".")[-1] == "org"

    return list(filter(check_domain_extension, strings))
```
